```python
us_df_clean = clean_names(us_df)
```

```python
us_df_clean.describe()
```

```python
import statsmodels.formula.api as smf
```

```python
x = us_df_clean[["change_in_google_stress_index"]]
```

```python
y = us_df_clean[["change_in_unemployment_rate"]]
```

```python
model = smf.ols(formula = "y ~ x", data = us_df_clean).fit()
```

```python
print(model.summary())
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
b, m = model.params
b,m
```

```python
plt.figure(figsize=(10,4.5))
plt.scatter(x, y, c = "b", s = 20)
plt.plot(x, model.fittedvalues, color ="r", linestyle = "-", linewidth = .5)
plt.text(1.28, 4.7, "Conneticut", ha = "center")
plt.text(-3.55, 0.7, "North Datkota", ha = "center")
plt.text(1.16, 9.1, "Nevada", ha = "center")
plt.text(-.44, 7, "California", ha = "center")
plt.text(-2.16, 4, "Virginia", ha="center")
plt.text(-4.16, 2.2, "South Dakota", ha="center")
plt.text(1.89, 4.6, "Kentucky")
plt.text(3.48, 5.7, "Idaho")
plt.text(3,4.5,"OLS line: y = " + "{:.2f}".format(b) + " + {:.2f}".format(m) + "x",
         size=10, c="purple",
        bbox = dict(facecolor = 'green', alpha = 0.1))
plt.xlabel("Change in Google search for antidepressants (2007-2010)")
plt.ylabel("Change in Unemployment (%,2007-2010)")
plt.title("Fig. 13.1 - The Core Ecconomics, Unity 13")
plt.axvline(x=0, c="y", linewidth = .5)
plt.axhline(y=4, c="y", linewidth = .5)
```
